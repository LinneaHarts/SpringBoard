```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as pp

%matplotlib inline
```

```python
tb_df = pd.read_csv('tb.csv')
tb_df.head()
```

```python
tb_df.columns
```

```python
tb_melt = pd.melt(tb_df, id_vars=['country', 'year'], value_vars=['m04', 'm514', 'm014', 'm1524', 'm2534', 'm3544',
       'm4554', 'm5564', 'm65', 'mu', 'f04', 'f514', 'f014', 'f1524', 'f2534',
       'f3544', 'f4554', 'f5564', 'f65', 'fu'], var_name='age_group', value_name='cases')
tb_melt
```

```python
tb_drop_na = tb_melt.dropna(subset=['cases'])
tb_drop_na
```

```python
tb_drop_na.loc[:]['sex'] = tb_drop_na.age_group.apply(lambda a: a[0])
tb_drop_na
```

```python
new_age_group = tb_drop_na['age_group'].map(lambda a: a[1:])
```

```python
tb_drop_na.loc[:]['age_group'] = new_age_group
tb_drop_na
```
